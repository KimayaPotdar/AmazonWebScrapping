```python
import io
import pandas as pd
data = pd.read_csv('Mask_reviews.csv',skipinitialspace=True)
```

```python
data
```

```python
data.comments = data.comments.str.strip(' \n\t')
data.review_title = data.review_title.str.strip(' \n\t')
```

```python
data
```

```python
import numpy as np 
from subprocess import check_output
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

word_string=" ".join(data['comments'].str.lower())
```

```python
word_string
```

```python
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = STOPWORDS,
                min_font_size = 11).generate(word_string) 
```

```python
# plot the WordCloud image                        
plt.figure(figsize = (8, 8)) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```
